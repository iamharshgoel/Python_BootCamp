```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('title')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,y)
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
ax2.plot(x,y)
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

#Large
ax1.plot(x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')

#Insert
ax2.plot(x,y)
ax2.set_title('zoom')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig,axes = plt.subplots(1,2)

axes[0].plot(x,y,color='blue',ls='--',lw=3)
axes[1].plot(x,z,color='red',lw=3)
```

```python
fig,axes = plt.subplots(1,2,figsize=(12,2))

axes[0].plot(x,y,color='blue',ls='--',lw=3)
axes[1].plot(x,z,color='red',lw=3)
```
